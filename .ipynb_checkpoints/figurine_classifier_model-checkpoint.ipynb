```python
import cv2 as cv
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv.imread('./test/sp1.jpg')
img.shape
```

```python
plt.imshow(img)
```

```python
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
gray.shape
```

```python
plt.imshow(gray, cmap='gray')
```

```python
face_cascade = cv.CascadeClassifier('./opencv/haar_face.xml')
eye_cascade = cv.CascadeClassifier('./opencv/haarcascade_eye.xml')
faces = face_cascade.detectMultiScale(gray,1.1,5)
faces
```

```python
(x,y,w,h) = faces[0]
x,y,w,h
```

```python
face = cv.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
plt.imshow(face)
```

```python
cv.destroyAllWindows()
for (x,y,w,h) in faces:
    face = cv.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = face[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,0,255),2)

plt.figure()
plt.imshow(face,cmap='gray')
plt.show()
```

```python
%matplotlib inline
plt.imshow(roi_color, cmap='gray')
```

```python
def get_cropped_img(img_path):
    img = cv.imread(img_path)
    gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray,1.1,5)
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes)>=2:
            return roi_color
```

```python
org = cv.imread('./test/sp1.jpg')
plt.imshow(org)
```

```python
cropped = get_cropped_img('./test/sp1.jpg')
plt.imshow(cropped)
```
